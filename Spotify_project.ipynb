```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

1. Importing data

```python
df_tracks = pd.read_csv('/Users/pranalidarekar/Downloads/tracks.csv')
df_tracks.head()
```

```python
# checking null values

pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
# top 10 least popular songs

sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
# Top 10 song with popularity greater than 90

most_popular = df_tracks.query('popularity > 90',inplace = False).sort_values('popularity', ascending= False)
most_popular[:10]
```

```python
# Making release date as our index

df_tracks.set_index("release_date",inplace= True)
df_tracks.index= pd.to_datetime(df_tracks.index, format='%Y-%m-%d', errors='coerce')
# Dates that cannot be parsed will use a date above them
df_tracks.index = pd.Series(df_tracks.index).fillna(method='ffill').values
df_tracks.head()
```

```python
# locating artist

df_tracks[["artists"]].iloc[18]
```

```python
# Coverting duration from MS to S
df_tracks['duration']= df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms",inplace= True, axis= 1)
```

```python
df_tracks.duration.head()
```

```python
# Correlation

corr_df= df_tracks.drop(['key','mode','explicit','id','name','artists','id_artists'],axis =1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df,annot= True, fmt= '.1g', vmin= -1,vmax= 1, center= 0, cmap ='inferno', linewidths= 1, linecolor= 'Black')
heatmap.set_title('Correlation Heatmap between variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
sample_df  = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(df_tracks))
```

```python
# creating a regression plot with loudness and energy

plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y= 'loudness', x= 'energy', color= 'c').set(title = 'Loudness vs Energy Correlation')
```

```python
# creating a regression plot with popularity and acousticness

plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y= 'popularity', x= 'acousticness', color= 'r').set(title = 'Popularity vs Acousticness Correlation')
```

```python
#df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
#df_tracks.dates = pd.to_datetime(df_tracks.dates)
#year= df_tracks.dates.dt.year
# Assuming the index itself contains the release dates
df_tracks['dates'] = df_tracks.index
df_tracks['dates'] = pd.to_datetime(df_tracks['dates'])
year = df_tracks['dates'].dt.year
df_tracks
```

```python
# creating a distribution plot number of songs each year since 1922
sns.displot(year, discrete= True, aspect= 2, height= 5, kind= 'hist').set(title= 'Number of song per year')
```

```python
# creating a bar plot to see duration of song over the year

total_dr= df_tracks.duration
fig_dm= (18,7)
fig, ax= plt.subplots(figsize= fig_dm)
fig= sns.barplot(x= year, y= total_dr,ax= ax, errwidth= False).set(title= "Duration vs Years")
plt.xticks(rotation= 90)
```
